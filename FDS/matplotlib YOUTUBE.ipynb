```python
x = np.linspace(0,10,15)
x
```

```python
plt.plot(x, np.sin(x), color = "orange", linestyle="dashed", label="sin(x)");
plt.plot(x, np.cos(x), color = "blue", label="cos(x)")
plt.title("TITLE")
plt.xlabel("Trục hoành")
plt.ylabel("sinx")


#giới hạn 1 khoảng của đồ thị
plt.xlim([0,4])
plt.ylim([-1,4])
#cach 2:
xmin=0
xmax=4
ymin=-1
ymax=4
plt.axis([xmin,xmax,ymin,ymax])

#để x với y nó fit vào biểu đồ luôn
plt.axis("tight")

#để x và y bằng nhau
plt.axis("equal")

plt.legend() #giúp các bạn chú thích những thành phần được vẽ trong biểu đồ -- label
```
