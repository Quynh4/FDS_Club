```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('Advertising.csv')
x = data.values[:, 2] #lấy tất cả các hàng của cột radio
y = data.values[:, 4] #cột sale
plt.scatter(x,y, marker='o')
#plt.show()
```

```python
# y = a*x + b
# m = weight
# x = radio
# b = bias
def predict(x,a, b):
    return a*x + b

def cost_function(x,y,a, b): #mảng x,y và giá trị a,b hiện tại
    n = len(x) #n là số phần tử của x
    sum_error = 0
    for i in range(n):
        sum_error += (y[i] - (a*x[i]+b))**2
    return  sum_error/n


def update_weight(x, y, a,b, learning_rate ):
    n= len(x)
    a_tempt = 0.0
    b_tempt = 0.0
    for i in range(n):
        a_tempt += -2*x[i]*(y[i] - (a*x[i]+b))
        b_tempt += -2*(y[i] - (a*x[i]+b))
    a -= (a_tempt/n)*learning_rate
    b -=(b/n)*learning_rate
    return  a,b 

def train(x, y, a,b, learning_rate, iter):
    cos_his = []
    for i in range(iter):
        a, b = update_weight(x, y, a,b, learning_rate)
        cost = cost_function(x,y, a, b)
        cos_his.append(cost)

    return a,b, cos_his

a,b, cos_his= train(x,y, 0.03, 0.0014, 0.001, 30)
print("ket qua:")
print(a)
print(b)
```
