```python
class km:
    def distance(self, pointA, pointB):
        output = 0
        for i in range(len(pointA)):
            output += (pointA[i] - pointB[i])**2
        return output ** .5
    
    def split_data(self, data, centroids):
        # generate matrix.
        matrix = []
        for sm_data in data:
            row = []
            for centroid in centroids:
                row.append(self.distance(sm_data, centroid))
            matrix.append(row)
        # from matrix create output.
        output = []
        for point in matrix:
            output.append(numpy.argmin(point))
        return numpy.array(output)
    
    def update(self, data, centroids, labels):
        new_centroids = []
        set_labels = [i for i in range(len(centroids))]
        for j in range(len(set_labels)):
            count = 0
            distance = numpy.array(centroids[j])
            for i in range(len(labels)):
                if set_labels[j] == labels[i]:
                    count += 1
                    distance += numpy.array(data[i])
            new_centroids.append(distance/count)
        return new_centroids
    
    def train(self, data, k, deep):
        self.data = data
        #chose k points
        centroids = random.choices(data, k = k)
        for _ in range(deep):
            labels = self.split_data(data, centroids)
            centroids = self.update(data, centroids, labels)
        return numpy.array(centroids)
```
