### Bài 1: đọc kiểu dữ liệu của dòng 3 (dòng có item_name là Chips and Tomatillo-Green Chili Salsa) cột 4(item_price) từ dataset 
### https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv

```python
data = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv')
type(data.item_price[3])

```

### Bài 2: chuyển dữ liệu của cột 4 (item_price) về float

```python
# nhớ xóa dấu $
data.item_price = data.item_price.str.replace('$', '')
data.item_price = data.item_price.astype('float32')
data.item_price.dtypes
```

### Bài 4: hiện bảng đã sort theo cột item_name và copy ra một bảng mới

```python
# Code ở đây
data1 = data.copy()
data1 = data1.sort_values('item_name')
data1
```

### Bài 5: chọn từ cột item_name đến hết item_price với các dòng từ 0 đến hết 4 và các dòng từ 4617 đến hết và lưu thành dataframe mới với tên là "Bài 5.csv"

```python
# Dùng to_csv
new_data = data.drop(columns = ['order_id', 'quantity'])
new_data = pd.concat([new_data[0:4], new_data[4617:]])
new_data.to_csv('Bai5.csv')
d = pd.read_csv('Bai5.csv')
d
```
