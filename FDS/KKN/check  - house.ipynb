```python
import numpy
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
data = pd.read_csv('house.csv')
target = data.condition
data = pd.concat([data.sqft_living, data.bathrooms, data.sqft_above, data.sqft_basement, data.price], axis=1)
data
```

```python
data = [[target[i],list(data[i])] for i in range(len(data))] # Mỗi phần tử trong list mới sẽ là một list gồm hai phần tử: phần tử đầu tiên là target[i] và phần tử thứ hai là list(data[i])

random.shuffle(data)

class knn:
    def train(self, data):
        self.data = data
        
    def distance(self, pointA, pointB):
        output = 0
        for i in range(len(pointA)):
            output += (pointA[i] - pointB[i])**2
        return output ** .5
    
    def predict(self, point, k):
        data = [val[1] for val in data]
        target = [val[0] for val in data]
        matrix = [[target[i],self.distance(data[i],point)] for i in range(len(data))]
        matrix.sort(key = lambda x: x[1])
        matrix = matrix[:k]
        matrix = [val[0] for val in matrix]
        
        output = []
        set_matrix = set(matrix)
        for val in set_matrix:
            count = 0
            for sm_val in matrix:
                if val == sm_val:
                    count += 1
            output.append([val,count/k])
        output.sort(key = lambda x: x[1], reverse=True)
        return output
    
```

```python
train_data = data[:4000]
test_data = data[4000:]
len(train_data),len(test_data)
```

```python

my_knn = knn()
my_knn.train(train_data)
k = 5
count = 0
for sm_data in test_data:
    if sm_data[0] == my_knn.predict(sm_data[1],k)[0][0]:
        count += 1
print('acc:',count/len(test_data) * 100,'%')
```
