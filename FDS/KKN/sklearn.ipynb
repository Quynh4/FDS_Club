```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas
import matplotlib.pyplot as plt
import random
```

```python
iris_data = load_iris()
target_names = iris_data.target_names
data = iris_data.data #numpy.array
target = iris_data.target #numpy.array_vector
feature_names = iris_data.feature_names #list
target_names = [target_names[val] for val in target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state =0 )
print(y_test)
print(type(X_train))
```

```python
model = DecisionTreeClassifier()
mymodel = model.fit(X_train, y_train)
print(mymodel.predict(X_test))
```

```python
X_new = np.array([[6.0, 3.23, 4.5, 2.0]])
print(mymodel.predict(X_new))
print(mymodel.score(X_test, y_test)) #do chinh xac
```
