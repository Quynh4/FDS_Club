```python
## Ở một thế giới khác -- Code thuật toán KNN
import csv
import numpy as np
import math
import random
from collections import Counter
```

```python
file = open("Iris.csv")
dataset = csv.reader(file)
dataset = np.array(list(dataset))
dataset = np.delete(dataset,0,0) #xoa hang 1
dataset = np.delete(dataset,0,1) #xoa cot 1
#xoa hang 2: np.delete(dataset,1,0)
#xoa cot 2: np.delete(dataset,1,1)
```

```python
random.shuffle(dataset) #trộn nó lên
print(dataset)
```

```python
#chia làm 2 phần
train = dataset[:120]
test = dataset[120:]
```

```python

def tinhKcach (a, b):
    result =0 
    #vì nó có 4 feature nên mk sẽ cho i chạy từ 1 đến 4
    # a và b là 2 cái điểm khác nhau, và mỗi điểm thì đều có 4 thuộc tính như thế tương ứng với các cột?
    # a[0] sẽ là 1 cái ô nào đấy, kiểu thế
    for i in range(4): 
        result+= (float(a[i]) - float(b[i]))**2
    
    return math.sqrt(result)
```

```python
def ganNhat(train, i, k):
    kcach = []
    #tinh kcach tu i den moi diem trong train
    for point in train:
        kcach.append(
            {
                "label": point[-1],
                "value": tinhKcach(i,point)
            })
        kcach.sort (key = lambda x:x["value"])
        labels= [i["label"] for i in kcach]
    return labels [:k]
        
```

```python
def voteTheDistance (knn):
    knn = ganNhat(train, i, k)
    label = Counter(knn).most_common()
    return label
```

```python
k = 5 #số các điểm
for i in test:
    knn = ganNhat(train, i, k) #tính khoảng cách giữa i đến all điểm trong trainset và trả về k
    result = voteTheDistance(knn)
    print("label:{} -> predict{}". format (i[-1], result))
```
