Cho số a và b, nếu a lớn hơn b in ra giá trị a - b, nếu không in ra giá trị b - a.

sample:

input: a = 2, b = 3 |
output: 1

input: a = 4, b = 1 |
output: 3

```python
def theGreater(a,b):
  if a>b:
    print(a-b)
  else:
    print(b-a)
a=int(input())
b=int(input())
theGreater(a,b)
```

Cho số nguyên n và chuỗi kí tự k bất kì, hãy in ra n lần chuỗi k.

sample:

input: n = 3, k = "hello"

output:

hello

hello

hello

```python
def nTimek(n,k):
  for i in range(n-1):
    print(k)
n=int(input())
k=input()
nTimek(n,k)


```

Cho một số n bất kì, hãy đưa ra tất cả các số lớn hơn 0, nhỏ hơn n và chia hết cho 3.

sample:

input: n = 4

output: 3

input: n = 9

output: 3 6

input: n = 14

output: 3 6 9 12

```python
def allDevineBy3(n):
  for i in range(1,n):
    if i % 3 == 0:
      print(i)
n=int(input())
allDevineBy3(n)
```
