```python
#about subplot
import matplotlib.pyplot as plt

fig = plt.figure()

# Subplot 1 - 1 hàng, 2 cột, vị trí 1
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot([1, 2, 3, 4], [1, 4, 2, 3])


# Subplot 2 - 1 hàng, 2 cột, vị trí 2
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot([1, 2, 3, 4], [4, 3, 2, 1])

plt.show()
```
